Q.2 How does broadcasting work in NumPy?

    Ans. Broadcasting in NumPy is a set of rules that lets you perform arithmetic operations on arrays of different shapes—as long as the shapes are compatible. Instead of throwing an error when shapes differ, NumPy tries to stretch (broadcast) the smaller array across the bigger one so the operation can still be performed element-wise.

    How it works-
    When operating on two arrays, NumPy compares their shapes element-wise, starting from the trailing dimensions (right to left).
    These are the rules:
    1- If the two dimensions are equal, they are compatible.
    2- If one of them is 1, it gets stretched to match the other.
    3- If they’re not equal and neither is 1, it throws an error.

Q.17 Explain the difference between apply() and map() in Pandas?

    Ans. Both apply() and map() are used to apply functions to Series or DataFrames. However, there are key differences in how and where each function is used. Let’s break down the differences:
    1- Functionality and Scope
    map():
    Used for Series only (1D).
    It applies a function element-wise on the individual elements of a Series.
    It’s typically used for transforming or mapping values in a Series.
    Supports passing dictionaries or series for direct mapping.

    apply():
    Can be used for both Series and DataFrames.
    It can apply a function along a specific axis (row-wise or column-wise).
    For DataFrames, it can apply a function to either rows (axis=1) or columns (axis=0).
    It provides more flexibility compared to map() as you can apply a function to each element or entire rows/columns of a DataFrame.

Q.21 Why is NumPy’s array slicing faster than Python’s list slicing?

    Ans. This is why NumPy is faster than Python

    1-Contiguous Memory Layout: NumPy arrays are stored in contiguous blocks of memory, making slicing more efficient.

    2- Memory Views: NumPy creates views instead of copies, reducing overhead in slicing.

    3-Low-level Optimizations: NumPy’s slicing is implemented in highly optimized C/Fortran code, leading to faster operations.

    4-Homogeneous Data Types: NumPy arrays are homogeneous, so operations can be optimized for specific data types.

    5-Efficient Memory Management: NumPy manages memory more efficiently, avoiding the overhead of copying data.
